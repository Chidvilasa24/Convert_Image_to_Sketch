generally our image is in RGB format but opencv uses BGR format So we are reversing the image(BGR reversed to RGB) get the correct image


```python
rev_image = img[:,:,::-1]
```

```python
def show_image(img_arr):
  rgb_arr = cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB)
  plt.imshow(rgb_arr)
  plt.axis(False)
```

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show_image(255-grey_img)
```

```python
grey_img
```

We have to make lighter ones darker aand draker ones lighter so we subtracting that from 255

```python
255-grey_img
```

We will a blur over this image now
BLUR IMAGE

```python
blur_image = cv2.GaussianBlur(255-grey_img, (111, 111), 0)
```

```python
show_image(blur_image)
```

```python
sketch = cv2.divide(blur_image, 255 - grey_img, scale = 256.0)
```

```python
show_image(sketch)
```
